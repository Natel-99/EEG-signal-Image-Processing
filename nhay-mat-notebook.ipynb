```python
def build_model(num_classes):
    model = tf.keras.models.Sequential()

    model.add(Conv1D(filters=32 , kernel_size=(3,), padding='same', activation=tf.keras.layers.LeakyReLU(alpha=0.001), input_shape = (14,256)))

    model.add(Conv1D(filters=64 , kernel_size=(3,), padding='same', activation=tf.keras.layers.LeakyReLU(alpha=0.001)))

    model.add(Conv1D(filters=128, kernel_size=(3,), padding='same', activation=tf.keras.layers.LeakyReLU(alpha=0.001)))

    model.add(Conv1D(filters=256, kernel_size=(3,), padding='same', activation=tf.keras.layers.LeakyReLU(alpha=0.001)))
    
    model.add(MaxPool1D(pool_size=3, strides=2, padding='same'))
    model.add(Dropout(0.5))

    model.add(Flatten())
    
    model.add(Dense(units = 512, activation=tf.keras.layers.LeakyReLU(alpha=0.001)))

    model.add(Dense(units = 256, activation=tf.keras.layers.LeakyReLU(alpha=0.001)))

    model.add(Dense(units = 128, activation=tf.keras.layers.LeakyReLU(alpha=0.001)))
    
    model.add(Dense(units = 64, activation=tf.keras.layers.LeakyReLU(alpha=0.001)))
    
    model.add(Dense(units = 32, activation=tf.keras.layers.LeakyReLU(alpha=0.001)))
   
    model.add(Dense(units = 16, activation=tf.keras.layers.LeakyReLU(alpha=0.001)))

    model.add(Dense(units = num_classes, activation='softmax'))
    
    adam = tf.keras.optimizers.Adam(learning_rate=0.00001)
    model.compile(optimizer=adam, 
                  loss = 'sparse_categorical_crossentropy', 
                  metrics=['accuracy'])
    return model
```

```python
class_model = build_model(num_classes=2)

class_history = class_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    batch_size=32,
    epochs=300,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        )
    ]
)
```

```python
class_model.summary()
```

```python
class_acc = class_model.evaluate(X_test, y_test, verbose=0)[1]
print("Test Accuracy (Class Model): {:.2f}%".format(class_acc * 100))
```

```python
import matplotlib.pyplot as plt
# summarize history for accuracy
plt.subplot(1,2,1)
plt.plot(class_history.history['accuracy'])
plt.plot(class_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
#plt.savefig('accuracy.png')
# summarize history for loss
plt.subplot(1,2,2)
plt.plot(class_history.history['loss'])
plt.plot(class_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
#plt.savefig('loss.png')

plt.show()
```
